```python
path = '/content/drive/MyDrive/TFM/nuevosdatos'
df_excel_01b  = pd.read_excel(path+'/EXCEL_01B.xlsx')
df_excel_04  = pd.read_excel(path+'/EXCEL_04.xlsx')
```

#### excel_01b

```python
# Select columns to identify each population centre
cols_identifier = df_excel_01b.columns[:2]
```

```python
# Select columns for male and female population
cols_men = df_excel_01b.columns[pd.Series(df_excel_01b.columns).str.startswith("HOM")]
cols_women = df_excel_01b.columns[pd.Series(df_excel_01b.columns).str.startswith("MUJ")]
```

```python
# Select sub-dataframe for male and female population
df_men_women = df_excel_01b[[*cols_identifier, *cols_men, *cols_women]]
```

#### excel_04

```python
# Select columns for age interval
cols_ages = [col for col in df_excel_04.columns if "_" in col]
```

```python
# Select subdataframe with CODMUN to identify each population centre
df_ages = df_excel_04[[*["CODMUN"], *cols_ages]]
```

## SELECCIÓN DE DATOS PARA 2010

```python
# Filter male and female population in 2010
df_men_women_2010 = df_men_women.filter(regex = "^.*?2010$|CODMUN|NOMBRE", axis = 1)
```

```python
# Filter age intervals for year 2010
df_ages_2010 = df_ages.filter(regex = "M.*?10$|H.*?10$|CODMUN$", axis = 1)
```

```python
# Merge both dataframes
df_2010 = reduce(lambda left,right: pd.merge(left,right,on='CODMUN'), [df_men_women_2010, df_ages_2010])
```
